## Ejercicio 1

Escribe una función que calcule el área de un triángulo, recibiendo la altura y la base como parámetros y otra función que calcule el área de un círculo recibiendo el radio del mismo.

```python
import numpy as np

def triangulo(base, alt):
    area = 0.5 * base * alt
    
    return print('El area del triangulo es: '+ str(area))

def circulo(r):
    area =  np.pi * (r ** 2)
    
    return print('El area del circulo es: '+ str(area))

triangulo(2,4)
circulo(2)
```

## Ejercicio 2

Escribe una función que pueda decirte si un número (número entero) es primo o no.

```python
def es_primo(z):
    for i in range(2, z):
        if (z % i ==0):
            return print(str(z) + ' no es primo, ya que '+ str(i) +' es divisor.')
    return print(str(z) + ' es primo')

for i in range(1,100):
    es_primo(i)
```

## Ejercicio 3

Escribe una función que pueda decirte si un año (número entero) es bisiesto o no.

```python
def bisiesto(year):
    if (year % 4 ==0) or ((year % 4 ==0) and (year % 100 ==0) and (year % 400 ==0)):
        return print('El año ' + str(year) + ' es bisiesto.')
    else:
        return print('El año ' + str(year) + ' no es bisiesto.')

for i in range(2000, 2050):
    bisiesto(i)
```
